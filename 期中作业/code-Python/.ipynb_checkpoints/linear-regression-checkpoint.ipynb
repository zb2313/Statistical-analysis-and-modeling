```python
import pandas as pd
import pretreat as pt
import corr
from sklearn.model_selection import train_test_split, KFold
from sklearn.utils import shuffle
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sa
from scipy import stats
from collections import Counter
```

```python
f1 = pt.get_data()
attributes = corr.get_attributes(f1)
f1_attr20 = f1[attributes]
# print(f1_attr20)
f1_attr20 = shuffle(f1_attr20)
```

```python
print(f1_attr20[attributes[:20]])
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(f1_attr20[attributes[:20]],f1_attr20['pIC50'],train_size=0.80)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

# 使用sklearn（梯度下降）

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
a  = model.intercept_#截距
b = model.coef_#回归系数
```

```python
print('----------------------------------')
print('      regression coefficient')
for i in range(0,20):
    print(attributes[i],b[i])
print('----------------------------------')
print('intercept',a)
print('----------------------------------')
```

```python
score = model.score(X_test,Y_test)
print(score)
```

```python
print(model.predict(X_test)[:200])
print(Y_test[:200])
```

```python
plt.figure(figsize=(5, 3), dpi=100)
plt.style.use('ggplot')
plt.plot(range(0,100), Y_test[:100],label="Y_test")
plt.legend(loc=4, fontsize=7)
plt.plot(range(0,100),model.predict(X_test)[:100],label="prediction")
plt.legend(loc=4, fontsize=7)
plt.title('overview of the model',fontsize=10)
plt.xlabel('index of data',fontsize=10)
plt.ylabel('pIC50',fontsize=10)
plt.show()
```

```python
mse=sm.mean_squared_error(Y_test,model.predict(X_test))#均方误差
rmse = np.sqrt(mse)#均方差根
evs=sm.explained_variance_score(Y_test, model.predict(X_test), sample_weight=None)#回归方差
me=sm.max_error(Y_test,model.predict(X_test))#最大误差
mae=sm.mean_absolute_error(Y_test,model.predict(X_test))#平均绝对误差回归损失
r2=sm.r2_score(Y_test,model.predict(X_test))#r2检验
print('-------------------------------------------')
print('                model overview        ')
print('mean squared error      ',mse)
print('root mean squared error ',rmse)
print('explained variance score',evs)
print('max error               ',me)
print('mean absolute error     ',mae)
print('R squared score         ',r2)
print('-------------------------------------------')
```

# 使用statsmodels(最小二乘法)

```python
model2=sa.OLS(Y_train,X_train)
```

```python
results = model2.fit() # 返回模型拟合结果
yFit = results.fittedvalues
print(results.summary())
```

```python
mse2=sm.mean_squared_error(Y_test,results.predict(X_test))#均方误差
rmse2 = np.sqrt(mse)#均方差根
evs2=sm.explained_variance_score(Y_test, results.predict(X_test), sample_weight=None)#回归方差
me2=sm.max_error(Y_test,results.predict(X_test))#最大误差
mae2=sm.mean_absolute_error(Y_test,results.predict(X_test))#平均绝对误差回归损失
r22=sm.r2_score(Y_test,results.predict(X_test))#r2检验
print('-------------------------------------------')
print('                model overview        ')
print('mean squared error      ',mse2)
print('root mean squared error ',rmse2)
print('explained variance score',evs2)
print('max error               ',me2)
print('mean absolute error     ',mae2)
print('R squared score         ',r22)
print('-------------------------------------------')
```
